# Dataset dos Dias de Interesse [27/10 -- 28/10]:

```python
Fault_Data=Data.loc["95759":"95997", ["Local Time","DFIG DC LinkVoltageActual Minimum V",
                                      "DFIG DC LinkVoltageActual Average V",
                                      'DFIG DC LinkVoltageActual Maximum V',
                                      'DFIG LSC Current Minimum A',
                                       'DFIG LSC Current Maximum A',
                                       'DFIG LSC Current Average A',
                                       'DFIG LSC Current Deviation A']]

                                     
```

# Plot Valores Médios [DC Link Voltage, DFIG LSC Current]:

```python
sns.set(rc={'figure.figsize':(32,15)})
sns.set_style("whitegrid")


plot = Fault_Data.plot(x="Local Time", y=["DFIG DC LinkVoltageActual Average V",
                                   "DFIG LSC Current Average A",
                                    'DFIG LSC Current Deviation A'],
                                           title='Average',fontsize=20)

plot.legend(loc='best', prop={'size': 20})

plt.xticks(rotation=45)

plt.savefig('sample.png')
```

# Plot Valores Mínimos [DC Link Voltage, DFIG LSC Current]:



```python
sns.set(rc={'figure.figsize':(32,15)})
sns.set_style("whitegrid")

plot = Fault_Data.plot(x="Local Time", y=["DFIG DC LinkVoltageActual Minimum V",
                                   "DFIG LSC Current Minimum A",
                                    'DFIG LSC Current Deviation A'],
                                    title='Minimum',fontsize=20)

plot.legend(loc='best', prop={'size': 20})                                           
plt.xticks(rotation=45)
```

# Plot Valores Máximos [DC Link Voltage, DFIG LSC Current]:

```python
sns.set(rc={'figure.figsize':(32,15)})
sns.set_style("whitegrid")
plot = Fault_Data.plot(x="Local Time", y=["DFIG DC LinkVoltageActual Maximum V",
                                   "DFIG LSC Current Maximum A",
                                    'DFIG LSC Current Deviation A'],
                                    title='Maximum',fontsize=20)

plot.legend(loc='best', prop={'size': 20})
plt.xticks(rotation=45)
```
